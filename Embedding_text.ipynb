```
import glob
import pandas as pd

textbase = "./imdb/aclImdb/test/"

# test data, read in pd.
testdf = pd.DataFrame(columns=["content","target"])
for i in glob.glob(textbase + "neg/*.txt"):
    with open(i,"r",encoding="utf-8") as f:
        content = f.read()
    s = pd.Series([content,0],index=["content","target"])
    testdf = testdf.append(s,ignore_index=True)
for i in glob.glob(textbase + "pos/*.txt"):
    with open(i,"r",encoding="utf-8") as f:
        content = f.read()
    s = pd.Series([content,1],index=["content","target"])
    testdf = testdf.append(s,ignore_index=True)
testdf.head()
```

```
trainbase = "./imdb/aclImdb/train/"

# train data, read in pd
traindf = pd.DataFrame(columns=["content","target"])
for i in glob.glob(trainbase + "neg/*.txt"):
    with open(i,"r",encoding="utf-8")as f:
        content = f.read()
    s = pd.Series([content,0],index=["content","target"])
    traindf = traindf.append(s,ignore_index=True)
for i in glob.glob(trainbase + "pos/*.txt"):
    with open(i,"r",encoding="utf-8") as f:
        content = f.read()
    s = pd.Series([content,1],index=["content","target"])
    traindf = traindf.append(s,ignore_index=True)
traindf.head()
```

```
# pre-building
from keras.preprocessing.text import Tokenizer
# 先創造你care的詞的詞典, 並限定2000個字(出現頻率高到低)的辭典(等一下每個詞都是2000維度的向量)
tokens = Tokenizer(num_words=2000)

# 從traindf挑選2000詞
tokens.fit_on_texts(traindf["content"])
```

```
# 到底2000個詞是哪一個詞(抓出現次數最多的)
sorted(tokens.word_counts.items(), key=lambda x:x[1] , reverse=True)
```

```
# take a look
tokens.word_index
```

```
# 把每一篇文章的每一個詞都轉化成為數字
train_seq = tokens.texts_to_sequences(traindf["content"])
test_seq = tokens.texts_to_sequences(testdf["content"])

# take a look
# 因為每一篇文章不一樣多, 出現最多詞的文章1496個詞, 其他文章就會補上NaN
pd.DataFrame(train_seq)
```

```
# 截長(後面截)補短(前面補0)到每一篇文章都是同樣詞數
from keras.preprocessing.sequence import pad_sequences
train_pad = pad_sequences(train_seq, maxlen=200)
test_pad = pad_sequences(test_seq, maxlen=200)

# take a look
# 如果你看到0, 補短所出現的結果
pd.DataFrame(train_pad)
```

```
# building
from keras.models import Sequential
from keras.layers import Dense, Flatten, Embedding, Dropout
model = Sequential()
'''

Embedding 可視為把字number seq轉成您要的詞向量。

'''
# input_dim  >>> 詞典是2000維度
# output_dim >>> care 的詞要幾個?
# input_length >>> 您有幾個詞要經過黑盒子?
model.add(Embedding(input_dim=2000, output_dim=64, input_length=200))
model.add(Flatten())
model.add(Dense(128, activation="relu"))

# 避免某些神經元一直主宰一切，故隨機挑掉0.2~0.5
model.add(Dropout(0.2))

model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```
# check
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```
# trianing
model.fit(train_pad, traindf["target"], 
          batch_size=200, epochs=3, 
          verbose=2, validation_split=0.1)
```

```
# 評估
model.evaluate(test_pad, testdf["target"])
```

```
# saving model
model.save("sentiment.h5")
```

```
# feelback
from keras.models import load_model
l = load_model("./sentiment.h5")
l.evaluate(test_pad, testdf["target"])
```

## RNN

```
from keras.models import Sequential
from keras.layers import Dense, Embedding, Dropout, SimpleRNN

# 此處不需要Flatten喔!!!
model = Sequential()
model.add(Embedding(input_dim=2000, output_dim=64, input_length=200))
model.add(SimpleRNN(64))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```
model.fit(train_pad, traindf["target"], 
          batch_size=200, epochs=3, 
          verbose=2, validation_split=0.1)
```

```
model.evaluate(test_pad, testdf["target"])
```
