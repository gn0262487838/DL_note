```
from keras.datasets.mnist import load_data 

(x_train, y_train), (x_test, y_test) = load_data()

# demo 
load_data()
```

```
# shape 看型態，共60000，28X28像素,另外一筆以此類推。
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# 資料預處理

### 題目預處理(normalize)

> keras在準備隨機那組weights，跟normalize那組比較搭。

* way1:  0~1

* way2: -1~1

> keras MLP 需要攤開資料成為一維。

* by reshape

```
# normalize-data
x_train_shape = x_train.reshape(60000, 784) / 255 
x_test_shape = x_test.reshape(10000, 784) / 255
# take a look
pd.DataFrame(x_train_shape[0])
```

### 標籤預處理(one-hot encoding)

> 希望標籤的樣子會是 [p0, p1, ... p9]

* ex. 5 : [0, 0, 0, 0, 0.8, 0.2, 0, 0, 0, 0]

> 已內建好函式轉換。

* by from keras.utils.np_utils import to_categorical 


```
from keras.utils.np_utils import to_categorical

y_train_cate = to_categorical(y_train)
y_test_cate = to_categorical(y_test)
# take a look
print("原來:",y_test[0])
print("整理過後:",y_test_cate[0])
```

# 建立模型
> Activation

* 中間:relu

* 最後一層看結果:

    1.sigmoid
    
    2.softmax


> 第一種

* 蛋糕支架(sequential)，堆上一層layers。

    layers:dense(全連接層)

```
from keras.models import Sequential
from keras.layers import Dense
mlp = Sequential()
# 第一層: input_dim >>> 告知輸入有幾個!!!
# 784 >>> 128 , 128 * 784=100352(神經元的連結) 再加上128個bias(偏差) 共100480 param !!!
mlp.add(Dense(128, activation="relu",input_dim=784))
# 128 >>> 10 , 10 * 128=1280(神經元的連結) 再加上10個bias(偏差) 共1290 param !!!
mlp.add(Dense(10,activation="softmax"))
# take a look
mlp.summary()
```

```
# 確定模型訓練方式(compile)...等因素。
# loss:
# 1.最後輸出一個:binary_crossentropy
# 2.最後輸出多個:categorical_crossentropy
# optimizer:梯度下降時的優化，基於動量(速度的累積)
mlp.compile(
    loss = "categorical_crossentropy",
    metrics = ["accuracy"],
    optimizer = "adam"
)
```

```
# training
# batch:看幾筆做一次梯度下降(修正)
# epochs:整份考古題要看幾遍。假如我現在設定10，訓練完後模型已定型，現在改成5就表示再訓練五次的意思，不是從頭訓練5次的意思喔!!!
# 10epochs 200batch: 60000 * 10 /200 = 3000(次)
mlp.fit(
   x_train_shape,
   y_train_cate,
   batch_size=200,
   epochs=12, 
   validation_split=0.1
)
# 參數verbose=2 可調整不要出現進度格。
```

```
# 評估
mlp.evaluate(x_test_shape,y_test_cate)
```

```
# 測試
# predict: 秀機率 ; predict_classes: 秀分類
pre = mlp.predict_classes(x_test_shape)
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, pre))
```

```
# graph
import numpy as np
# numpy才能每個數字比較
# list: list[一個數字]
# numpy array: nparray[[多個數字]]
diff = np.nonzero(pre != y_test)[0]
print(y_test)
print(diff)
# 預測錯誤的圖的正確答案。
origin = y_test[diff]
# 預測錯誤的圖的錯誤答案。
fal = pre[diff]
# 把圖取出來
photo = x_test[diff]

plt.figure(figsize=(20,45))

width = 10
height = int(len(photo) / 10) + 1
for (i,j) in enumerate(photo):
    # 列數,行數, 第幾張圖(i開始)
    plt.subplot(height, width, i+1)
    plt.axis("off")
    t = "PRE:" + str(fal[i]) + " COR:" + str(origin[i])
    plt.title(t)
    plt.imshow(j, cmap="gray")
```
